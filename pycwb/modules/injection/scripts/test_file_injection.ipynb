```python
import sys 
import matplotlib.pyplot as plt
sys.path.insert(0,'..')
import numpy as np 
from inj_generators import get_strain_from_file
%load_ext autoreload



hdf_files = {'H1': 'injection_files/H1_150914.hdf',
             'L1': 'injection_files/L1_150914.hdf'}

npy_files = {'H1': 'injection_files/H1_150914.npy',
             'L1': 'injection_files/L1_150914.npy'} 

txt_files = {'H1': 'injection_files/H1_150914.txt',
             'L1': 'injection_files/L1_150914.txt'}

parameters_hdf = {
    'files': hdf_files,
    'gps_time': 1126259462,  # GPS time of the event
} 

parameters_txt = {
    'files': txt_files,
    'gps_time': 1126259462,  # Start time of the segment

} 

parameters_npy = {
    'files': npy_files,
    'gps_time': 1126259462,  # Start time of the segment
}

strain_hdf = get_strain_from_file(**parameters_hdf)
strain_txt = get_strain_from_file(**parameters_txt)
strain_npy = get_strain_from_file(**parameters_npy)


print(strain_hdf)
print(strain_txt)
print(strain_npy)
```

```python

fig = plt.figure(figsize=(8, 10))
for i, strain in enumerate([strain_hdf, strain_txt, strain_npy]):
    ax = fig.add_subplot(3, 1, i + 1)
    strain['H1'].plot()
```
